```python
import yfinance as yf
import pandas as pd

ticker = yf.Ticker("AAPL")
data = yf.download("AAPL", start="2020-01-01", end="2025-03-15")
print(data.tail())

data.to_csv("apple_stock_data.csv")

import matplotlib.pyplot as plt

data['Close'].plot(title="Apple Stock Closing Prices", figsize=(10, 6))
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.show()


```
